Python 문자열의 `isdigit()`과 `isalpha()` 메소드에 대해 설명해드리겠습니다.

## isdigit() 메소드

**기능:** 문자열이 모두 숫자로만 구성되어 있는지 확인

**문법:** `string.isdigit()`

**반환값:** 
- `True` - 모든 문자가 숫자인 경우
- `False` - 숫자가 아닌 문자가 하나라도 있거나 빈 문자열인 경우

```python
# 예시
print("123".isdigit())      # True
print("0".isdigit())        # True
print("123a".isdigit())     # False (문자 포함)
print("-123".isdigit())     # False (음수 기호 포함)
print("12.3".isdigit())     # False (소수점 포함)
print("".isdigit())         # False (빈 문자열)
print(" 123 ".isdigit())    # False (공백 포함)
```

## isalpha() 메소드

**기능:** 문자열이 모두 알파벳(영문자)으로만 구성되어 있는지 확인

**문법:** `string.isalpha()`

**반환값:**
- `True` - 모든 문자가 알파벳인 경우
- `False` - 알파벳이 아닌 문자가 하나라도 있거나 빈 문자열인 경우

```python
# 예시
print("abc".isalpha())      # True
print("ABC".isalpha())      # True
print("Hello".isalpha())    # True
print("안녕".isalpha())      # True (한글도 알파벳으로 인식)
print("abc123".isalpha())   # False (숫자 포함)
print("Hello World".isalpha()) # False (공백 포함)
print("".isalpha())         # False (빈 문자열)
print("Hello!".isalpha())   # False (특수문자 포함)
```

## 실제 활용 예시

```python
# 사용자 입력 검증
user_input = input("숫자를 입력하세요: ")
if user_input.isdigit():
    number = int(user_input)
    print(f"입력된 숫자: {number}")
else:
    print("올바른 숫자를 입력해주세요.")

# 이름 검증
name = input("이름을 입력하세요: ")
if name.isalpha():
    print(f"안녕하세요, {name}님!")
else:
    print("이름에는 문자만 입력해주세요.")
```

## 주의사항
- **isdigit()**: 음수나 소수는 `False` 반환
- **isalpha()**: 공백이나 특수문자가 포함되면 `False` 반환
- 둘 다 빈 문자열에 대해서는 `False` 반환
