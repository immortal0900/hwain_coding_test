Python에서 대소문자를 변환하는 주요 문자열 메소드들을 설명해드리겠습니다.

lower() = casefold() = casefold()는 특수한 유니코드 문자도 소문자로
upper()
title() = 각 단어의 첫글자 대문자로, 단어 하나일때도 가능
capitalize() = 가장 첫글자만 대문자로 나머지는 소문자로
정도는 기억해 두는 것이 좋을거 같음
## 대문자 → 소문자: lower()

**기능:** 모든 대문자를 소문자로 변환

```python
text = "HELLO WORLD"
result = text.lower()
print(result)  # "hello world"

# 예시
print("Python".lower())      # "python"
print("ABC123".lower())      # "abc123"
print("Hello World!".lower()) # "hello world!"
```

## 소문자 → 대문자: upper()

**기능:** 모든 소문자를 대문자로 변환

```python
text = "hello world"
result = text.upper()
print(result)  # "HELLO WORLD"

# 예시
print("python".upper())      # "PYTHON"
print("abc123".upper())      # "ABC123"
print("Hello World!".upper()) # "HELLO WORLD!"
```

## 기타 유용한 대소문자 변환 함수들

### 1. swapcase() - 대소문자 서로 바꾸기
```python
text = "Hello World"
print(text.swapcase())  # "hELLO wORLD"
```

### 2. capitalize() - 첫 글자만 대문자
```python
text = "hello world"
print(text.capitalize())  # "Hello world"
```

### 3. title() - 각 단어의 첫 글자를 대문자로
```python
text = "hello world python"
print(text.title())  # "Hello World Python"
```

### 4. casefold() - 더 강력한 소문자 변환
```python
# 특수한 유니코드 문자까지 고려한 소문자 변환
text = "ß"  # 독일어 에스체트
print(text.lower())     # "ß"
print(text.casefold())  # "ss"
```

## 실제 활용 예시

```python
# 사용자 입력을 대소문자 구분 없이 비교
user_input = input("Yes 또는 No를 입력하세요: ")
if user_input.lower() == "yes":
    print("네!")
elif user_input.lower() == "no":
    print("아니오!")

# 이름 정리 (첫 글자만 대문자)
name = input("이름을 입력하세요: ").strip().title()
print(f"안녕하세요, {name}님!")

# 패스워드 대소문자 검증
password = "MyPassword123"
has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
print(f"대문자 포함: {has_upper}")  # True
print(f"소문자 포함: {has_lower}")  # True
```

## 중요한 특징
- **원본 문자열은 변경되지 않음** (문자열은 불변객체)
- **새로운 문자열을 반환**
- **숫자나 특수문자는 영향받지 않음**
- **한글 등 다른 언어는 영향받지 않음**

```python
original = "Hello"
converted = original.lower()
print(original)   # "Hello" (원본 그대로)
print(converted)  # "hello" (새로운 문자열)
```
